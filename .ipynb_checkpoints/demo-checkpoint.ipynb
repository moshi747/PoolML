```python
import Vision.vision as vision
from pool import PoolAI, balls_to_obs, rescale, view_shots
import xgboost as xgb
import cv2
import matplotlib.pyplot as plt
import copy
```

```python
param = {'max_depth': 30, 'eta': 0.4, 'lambda': 100, 'objective': 'reg:squarederror'}

boosted = xgb.XGBRegressor(param)

boosted.load_model('boosted.json')

treemodel = PoolAI(model = boosted, features = [balls_to_obs], output = 'hit', hit_scale = 'raw')
```

```python
img_orig = cv2.imread('Vision/mytable2.jpeg')
h = vision.get_homography(img_orig, 'BLUE')
t_coords, o_coords = vision.get_coords(img_orig, h)
```

```python
img = copy.copy(img_orig)


for i, center in enumerate(o_coords):
    cv2.circle(img, center, 30, (0, 255, 0), 5)
    cv2.putText(img, text=f'{i}', org=(center[0]+20,center[1]+20),
                color=(0,255,255), fontFace=cv2.FONT_HERSHEY_DUPLEX, fontScale=3, thickness=3)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
t_coords = vision.reorder(t_coords, [2,4,5,6,1,8,3,0,9,7])
o_coords = vision.reorder(o_coords, [2,4,5,6,1,8,3,0,9,7])
```

```python
img = copy.copy(img_orig)


for i, center in enumerate(o_coords):
    cv2.circle(img, center, 30, (0, 255, 0), 5)
    cv2.putText(img, text=f'{i}', org=(center[0]+20,center[1]+20),
                color=(0,255,255), fontFace=cv2.FONT_HERSHEY_DUPLEX, fontScale=3, thickness=3)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
img = copy.copy(img_orig)

img = cv2.warpPerspective(img, h, (320,640))

for center in t_coords:
    cv2.circle(img, center, 30, (0, 255, 0), 5)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
for i in range(len(t_coords)):
    t_coords[i]=list(t_coords[i])

t_coords = rescale(t_coords, (320, 640))
```

```python
view_shots(treemodel, t_coords)
```
